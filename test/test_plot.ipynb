```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.backends.backend_pdf import PdfPages
```

```python
f = pd.read_excel('tests.xlsx')
```

```python
n = np.arange(1,f.shape[0]+1)
```

```python
fig_size = plt.rcParams["figure.figsize"]
fig_size[0] = 16
fig_size[1] = 16
plt.rcParams["figure.figsize"] = fig_size
```

```python
for col in f:
    plt.scatter(n, f[col], label='')
    line_style = '-' if col[1]=='1' else '--'
    plt.plot(n, f[col], linewidth=3, linestyle=line_style)

plt.xticks(n)
plt.yticks(np.arange(0, f.values.max()+1, 1))

plt.xlabel('# of measurement')
plt.ylabel('time [s]')
plt.legend(loc='upper right', ncol=3)

plt.grid(b=True, which='major', linestyle='-', color='black')
plt.grid(b=True, which='minor', linestyle='--')
plt.minorticks_on()
```

```python
pdf_page = PdfPages('tests.pdf')
plt.savefig(pdf_page, format='pdf')
pdf_page.close()
```
